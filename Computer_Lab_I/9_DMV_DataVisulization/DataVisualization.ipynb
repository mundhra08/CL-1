```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("city_day.csv")
```

```python
data = data.dropna()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
date_column = 'Date'
aqi_column = 'AQI'
pollutant_columns = ['PM2.5', 'PM10', 'CO','SO2','O3']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[date_column], data['PM2.5'], label='pollutant')
plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Trends Over Time')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[date_column], data['PM10'], label='pollutant')
plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Trends Over Time')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[date_column], data['CO'], label='pollutant')
plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Trends Over Time')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[date_column], data['SO2'], label='pollutant')
plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Trends Over Time')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[date_column], data['O3'], label='pollutant')
plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Trends Over Time')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([data[pollutant] for pollutant in pollutant_columns], labels= pollutant_columns)
plt.xlabel('Pollutant')
plt.ylabel('AQI Values')
plt.title('AQI Distribution for Different Pollutants')
plt.grid(True)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['PM2.5'], data[aqi_column], label='PM2.5', alpha=0.5)
plt.scatter(data['PM10'], data[aqi_column], label='PM10', alpha=0.5)
plt.scatter(data['CO'], data[aqi_column], label='CO', alpha=0.5)
plt.xlabel('Pollutant Levels')
plt.ylabel('AQI')
plt.title('Relationship Between AQI and Pollutants')
plt.legend()
plt.grid(True)
```
