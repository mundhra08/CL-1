```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Mumbai_Property.csv")
```

```python
data.columns = data.columns.str.replace(' ', '_')
data.columns = data.columns.str.replace('[^a-zA-Z0-9_]', '')
```

```python
data.isna().sum()
```

```python
filtered_data = data[(data['Price_Lakh'] >= 2000) & (data['Availability'] == 'Ready To Move')]
```

```python
filtered_data
```

```python
data = pd.get_dummies(data, columns=['Property_Age'], prefix=['Property_Age'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['Availability'], prefix=['Availability'])
```

```python
data.head()
```

```python
neighborhood_avg_prices = data.groupby('Area_Tpye')['Price_Lakh'].mean()
```

```python
neighborhood_avg_prices
```

```python
from scipy import stats
z_scores = np.abs(stats.zscore(data['Price_Lakh']))
data_no_outliers = data[(z_scores < 3)]
```

```python
data_no_outliers
```

```python
data.to_csv('Cleaned_Mumbai_RealEstate_Data.csv', index=False)
```
