```python
import pandas as pd
```

```python
csv_data = pd.read_csv(r"sales_data_sample.csv")
```

```python
excel_data = pd.read_excel("sales_data_sample.xlsx", sheet_name="Sheet1")
```

```python
with open("csvjson.json","r") as json_file:
    json_data = pd.read_json(json_file)
```

```python
unified_data = pd.concat([csv_data,excel_data,json_data],ignore_index=True)
```

```python
unified_data.info()
```

```python
unified_data.isnull().sum()
```

```python
unified_data.drop(['ADDRESSLINE2'],inplace= True, axis=1)
```

```python
unified_data.fillna("unknown",inplace=True)
```

```python
unified_data.isnull().sum()
```

```python
unified_data["TotalAmount"] = unified_data["QUANTITYORDERED"]*unified_data["PRICEEACH"]
```

```python
total_sales = unified_data["TotalAmount"].sum()
total_sales
```

```python
average_order_value = unified_data.groupby("ORDERNUMBER")["TotalAmount"].sum().mean()
average_order_value
```

```python
product_category_distribution = unified_data["PRODUCTLINE"].value_counts()
product_category_distribution
```

```python
import matplotlib.pyplot as plt
product_category_distribution.plot(kind = "bar")
plt.title("Product Category Distribution")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(product_category_distribution,labels= product_category_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title("Product_Ctegory_Distribution")
plt.show()
```

```python
unified_data["ORDERDATE"] = pd.to_datetime(unified_data["ORDERDATE"])

# Group by date and calculate total sales per day
daily_sales = unified_data.groupby(unified_data["ORDERDATE"].dt.date)["TotalAmount"].sum()

# Sort the data by date (optional but recommended)
daily_sales = daily_sales.sort_index()

# Create a line chart for daily sales over time
plt.figure(figsize=(10, 6))
plt.plot(daily_sales.index, daily_sales.values, marker='o', linestyle='-', color='b')
plt.title("Daily Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)  # Add a grid for better readability
plt.show()
```

```python
# List of product categories
product_categories = unified_data["PRODUCTLINE"].unique()

# Create an empty list to store data for each category
data_by_category = []

# Populate the data_by_category list with the TotalAmount for each category
for cat in product_categories:
    data_by_category.append(unified_data[unified_data["PRODUCTLINE"] == cat]["TotalAmount"])

# Create a boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data_by_category, labels=product_categories)
plt.title("Sales Distribution by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
 # Create a scatter plot to show the relationship between Price and Quantity
plt.figure(figsize=(8, 6))
plt.scatter(unified_data["PRICEEACH"], unified_data["QUANTITYORDERED"], alpha=0.5)
plt.title("Price vs. Quantity")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.show()
```

```python
# Convert date column to datetime
unified_data["OREDERDATE"] = pd.to_datetime(unified_data["ORDERDATE"])
# Extract month and year
unified_data["Month"] = unified_data["ORDERDATE"].dt.strftime("%Y-%m")
# Calculate total sales per month
monthly_sales = unified_data.groupby("Month")["TotalAmount"].sum()
# Create a bar plot for sales by month
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind="bar", color="steelblue")
plt.title("Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
